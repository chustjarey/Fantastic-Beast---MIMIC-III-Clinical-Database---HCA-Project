```python
#import libraries
import os
import gzip
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# reading admission file
file ='C:/Users/zuoch/Downloads/Final semester/DAB304- Health analytics/mimic-iii-clinical-database-1.4/ADMISSIONS.csv.gz'
admissions = pd.read_csv(file, compression ='gzip')
admissions
```

```python
# Display basic information about the dataset
admissions.info()
```

```python
# Check for missing values
print("\nMissing Values:")
print(admissions.isnull().sum())
```

```python
#replace the missing value 

admissions['DEATHTIME'].fillna('Unknown', inplace=True)
admissions['LANGUAGE'].fillna('Unknown', inplace=True)
admissions['RELIGION'].fillna('Unknown', inplace=True)
admissions['MARITAL_STATUS'].fillna('Unknown', inplace=True)
admissions['EDREGTIME'].fillna(0, inplace=True)
admissions['EDOUTTIME'].fillna(0, inplace=True)
admissions['DIAGNOSIS'].fillna(0, inplace=True)
```

```python
print(admissions.isnull().sum())
```

```python
admissions.columns
```

```python
#checking number of the columns and rows
admissions.shape
```

```python
# Display summary statistics for numerical columns
numeric_stats = admissions.describe()
numeric_stats
```

```python
# Time-related Analyses
admissions['ADMITTIME'] = pd.to_datetime(admissions['ADMITTIME'])
admissions['DISCHTIME'] = pd.to_datetime(admissions['DISCHTIME'])
admissions['LOS'] = (admissions['DISCHTIME'] - admissions['ADMITTIME']).dt.total_seconds() / (24 * 60 * 60)  # Length of Stay in days
admissions['ADMITTIME'] 
admissions['DISCHTIME']
admissions['LOS'] 
```

```python
# Correlation Matrix
numerical_columns = admissions.select_dtypes(include=['int64', 'float64']).columns
correlation_matrix = admissions[numerical_columns].corr()
correlation_matrix
```
